## Data Processing:


```python
import pandas as pd
```

```python
data = pd.read_excel('/content/Online retail.xlsx')
data
```

```python
data = data[data.columns[0]].str.split(',',expand = True)
data.head()
```

```python
data[data.duplicated()]
```

```python
data = data.drop_duplicates()
data
```

```python
data.isna()
```

```python
data.dropna(axis=1,how='all',inplace=True)
data.dropna(axis=0,how='all',inplace=True)
data
```

```python
cleaned_data = data.fillna('')
cleaned_data.head()
```

```python
transactions = cleaned_data.apply(lambda row: [item for item in row if item !=''],axis=1).tolist()
transactions[:5]
```

####Association Rule Mining:

```python
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te
```

```python
te_ary = te.fit(transactions).transform(transactions)
te_ary
```

```python
data = pd.DataFrame(te_ary,columns=te.columns_)
data.head()
```

```python
data.to_csv('data.csv',index=False)
```

```python
frequent_itemsets = apriori(data,min_support=0.01,use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules = association_rules(frequent_itemsets,metric='confidence',min_threshold=0.5)
rules
```

####Analysis and Interpretation:

```python
def format_itemset(itemset):
  return ','.join(sorted(list(itemset)))
```

```python
rules['antecedents_formatted'] = rules['antecedents'].apply(format_itemset)
rules['consequents_formatted'] = rules['consequents'].apply(format_itemset)
rules
```

```python
print("Top 10 Association Rules:")
rules[['antecedents_formatted','consequents_formatted','support','confidence','lift']]
```

```python
antecedents = rules['antecedents_formatted'].value_counts().head(5)
consequents = rules['consequents_formatted'].value_counts().head(5)
```

```python
print('Top 5 Antecedents:')
antecedents
```

```python
print('Top 5 Consequents:')
consequents
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.histplot(rules['lift'], bins=20, kde=True)
plt.title('Distribution of Lift Values')
plt.xlabel('Lift')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(rules['support'], rules['confidence'], alpha=0.5, color='blue')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
high_lift_rules = rules[(rules['lift'] > 1.5) & (rules['support'] > 0.01)]
high_lift_rules[['antecedents_formatted', 'consequents_formatted', 'support', 'confidence', 'lift']]
```

```python
rules['antecedent_size'] = rules['antecedents'].apply(len)
rules['consequent_size'] = rules['consequents'].apply(len)
```

```python
avg_antecedent_size = rules['antecedent_size'].mean()
avg_consequent_size = rules['consequent_size'].mean()
```

```python
print(f"Average antecedent size: {avg_antecedent_size:.2f}")
```

```python
print(f"Average consequent size: {avg_consequent_size:.2f}")
```

```python
all_products = rules['antecedents'].apply(list) + rules['consequents'].apply(list)
product_counts = pd.Series([item for sublist in all_products for item in sublist]).value_counts()
product_counts
```

Observations

High Lift Values: The rules with high lift values indicate strong associations between the products. For example, the combination of "frozen vegetables" and "ground beef" with "spaghetti" has a high lift, suggesting that these items are often purchased together.

Common Products: "Mineral water" frequently appears as a consequent, indicating it is a common follow-up purchase with other items.

Rule Sizes: The average antecedent size is 2, and the average consequent size is 1, indicating that most rules involve two products leading to one consequent product.



###Interview Questions:

####1.What is lift and why is it important in Association rules?

--> Lift,a key concept in association rule mining,measures the strength of an association rule.

--> It indicates how much greater the probability of the right-hand side(consequent) is in the presence of the lsft-hand side(antecedent) compared to its probability in the absence of the antecedent.

--> A higher lift value represents a more significant association,making it more important in decision-making processes.

--> Lift helps identifying the most relevant relationships between items,allowing bussinesses to optimize marketing stratagies,product recommendations,and customer segmentations.

####2.What is support and confidence.How do you calculate them?

--> Support and confidence are fundamental concepts in decision-making and machine learning,particular in the  classification and regression tasks.

--> Support refers to the proportion of instances in the training dataset that belongs to certain class or category.

--> Confidence represents the certainty or reliability of the model's prediction.

Calculation:

Support:Counting the number of instances in each class and dividing by the total number of instances.

Confidence:Calculated by dividing the number of transactions containing both itemsets by the number of transactions containing the first itemset.
