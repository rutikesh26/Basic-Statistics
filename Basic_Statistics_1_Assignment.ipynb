```python
# For Volume Mean,Medain , Mode
mean=round(sales['Volume'].mean(),2)
print("Mean of the sales for Volume is :",mean)

median=sales['Volume'].median()
print("Median of the sales for Volume is :",median)

mode=sales['Volume'].mode()
print("Mode of the sales for Volume is :",mode)

std=sales['Volume'].std()
print("Standard deviation of the sales for Volume is :",std)
```

**Box Plot**

```python
plt.figure(figsize=(8,6))
plt.boxplot([sales_num['Volume'], sales_num['Avg Price'], sales_num['Total Sales Value'],sales_num["Discount Rate (%)"],sales_num["Discount Amount"],sales_num["Net Sales Value"]], labels=['Volume', 'Avg Price', 'Total Sales Value',"Discount Rate (%)","Discount Amount","Net Sales Value"])
plt.title("Boxplot of Numerical Columns")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.xticks(rotation=45,ha="right")
plt.show()
```

Interquartile Range (IQR)

```python
columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
for col in columns:
    Q1 = sales_num[col].quantile(0.25)
    Q3 = sales_num[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = sales_num[col][(sales_num[col] < lower_bound) | (sales_num[col] > upper_bound)]

    print(f"Column: {col}")
    print(f"First Quartile (Q1): {Q1}")
    print(f"Third Quartile (Q3): {Q3}")
    print(f"Interquartile Range (IQR): {IQR}")
    print(f"Lower Bound for Outliers: {lower_bound}")
    print(f"Upper Bound for Outliers: {upper_bound}")
    print(f"Outliers in the '{col}' data: {outliers}\n")
```

Standardization of Numerical Variables

```python
sales_standarlized=sales.copy()
```

```python
for columns in sales.select_dtypes(include="number").columns:
  mean=sales[columns].mean()
  std=sales[columns].std()

  sales_standarlized[columns]=(sales[columns]-mean)/std
```
